```python
import pandas as pd
import numpy as np 
```

```python
RawData = pd.ExcelFile('BookData.xlsx')

dfs = {sheet_name: RawData.parse(sheet_name) 
          for sheet_name in RawData.sheet_names}
```

```python
Total_df = {}
for i in dfs.keys():
    temp = str(i+'_df')
    Total_df[temp] = pd.DataFrame(dfs[i].iloc[0:,1:11])
    Total_df[temp].columns = ['BookName','AuthorName','Edition','CopiesNumber','Publisher','PublicationYear','Dummy1','Dummy2','Dummy3','Address']
```

```python
temp = Total_df['Soft_df']
temp.drop(temp.index[24],inplace = True)
temp['Edition'] = temp['Edition'].fillna('NA')
temp['PublicationYear'] = temp['PublicationYear'].fillna('NA')
Total_df['Soft_df'] = temp
```

```python
temp = Total_df['ROB_df']
temp.drop(temp.index[7],inplace = True)
temp['Edition'] = temp['Edition'].fillna('NA')
Total_df['ROB_df'] = temp
```

```python
temp = Total_df['MP_df']
temp.drop(temp.index[0],inplace = True)
temp.drop(temp.index[17],inplace = True)
temp['Edition'] = temp['Edition'].fillna('NA')
Total_df['MP_df'] = temp
```

```python
temp = Total_df['EDC_df']
temp.drop(temp.index[5],inplace = True)
temp.drop(temp.index[11],inplace = True)
temp.drop(temp.index[12],inplace = True)
temp.drop(temp.index[17],inplace = True)
temp.drop(temp.index[17],inplace = True)
temp.drop(temp.index[20],inplace = True)
temp.drop(temp.index[25],inplace = True)
temp['Edition'] = temp['Edition'].fillna('NA')
temp['Publisher'] = temp['Publisher'].fillna('NA')
Total_df['EDC_df'] = temp
```
